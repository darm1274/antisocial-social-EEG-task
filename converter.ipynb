```python
import os
from moviepy.editor import VideoFileClip
```

```python
ruta_todos = './AUDIOS_TODOS'
carpeta_todos = os.listdir(ruta_todos)
```

```python
for carpeta in carpeta_todos:
    # Ruta de la carpeta de origen y destino
    carpeta_origen = f'./AUDIOS_TODOS/{carpeta}/AUDIOS'
    carpeta_destino = './AUDIOS'



    # Asegurarse de que la carpeta de destino existe, si no, crearla
    os.makedirs(carpeta_destino, exist_ok=True)

    # Iterar sobre todos los archivos en la carpeta de origen
    for archivo in os.listdir(carpeta_origen):
        # Comprobar si el archivo es un .mp4
        if archivo.endswith('.mp4'):
            ruta_video = os.path.join(carpeta_origen, archivo)
            # Crear el nombre del archivo de audio
            nombre_audio = os.path.splitext(archivo)[0] + '.mp3'
            ruta_audio = os.path.join(carpeta_destino, nombre_audio)
            
            # Convertir el archivo de video a audio
            with VideoFileClip(ruta_video) as video:
                video.audio.write_audiofile(ruta_audio)

    print("Conversión completada.")
```

```python
import os

# Ruta de la carpeta que contiene los archivos
carpeta = './AUDIOS'

# Iterar sobre todos los archivos en la carpeta
for archivo in os.listdir(carpeta):
    ruta_completa = os.path.join(carpeta, archivo)
    # Asegurarse de que es un archivo (no una carpeta)
    if os.path.isfile(ruta_completa):
        # Separar el nombre del archivo y su extensión
        nombre, extension = os.path.splitext(archivo)
        # Encontrar la posición del primer espacio en el nombre
        if ' ' in nombre:
            nuevo_nombre = nombre.split(' ')[0] + extension
            nueva_ruta = os.path.join(carpeta, nuevo_nombre)
            # Renombrar el archivo
            os.rename(ruta_completa, nueva_ruta)
            print(f'Renombrado: {archivo} -> {nuevo_nombre}')

print("Renombrado completado.")
```

```python
import pandas as pd

# Cargar el archivo Excel
archivo_excel = './4condiciones_160escenarios_EVA.xlsx'
xls = pd.ExcelFile(archivo_excel)

# Nombres de las hojas
hojas = ['SOCIAL ALINEADA', 'SOCIAL DESALINEADA', 'ANTISOCIAL ALINEADA', 'ANTISOCIAL DESALINEADA']

# Función para generar las rutas de los audios
def generar_ruta_audio(socialidad, parrafo, genero):
    if pd.isna(socialidad):
        return None  # O puedes devolver un valor por defecto si prefieres
    print(parrafo)
    return f"./AUDIOS/A{socialidad[0]}{genero}{int(parrafo)}.mp3"

# Procesar cada hoja
for hoja in hojas:
    # Cargar la hoja en un DataFrame
    df = pd.read_excel(archivo_excel, sheet_name=hoja)
    
    # Crear las nuevas columnas
    df['AUDIOS_M'] = df.apply(lambda row: generar_ruta_audio(row['Socialidad'], row['Párrafo'], 'M'), axis=1)
    df['AUDIOS_F'] = df.apply(lambda row: generar_ruta_audio(row['Socialidad'], row['Párrafo'], 'F'), axis=1)
    
    # Guardar los cambios en el archivo Excel
    with pd.ExcelWriter(archivo_excel, mode='a', if_sheet_exists='replace', engine='openpyxl') as writer:
        df.to_excel(writer, sheet_name=hoja, index=False)

print("Las columnas AUDIOS_M y AUDIOS_F han sido agregadas y llenadas correctamente.")
```
